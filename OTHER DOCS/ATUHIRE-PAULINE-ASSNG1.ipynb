```python
# Generate Fibonacci sequence of length 5
def fibonacci(n):
    seq = [0, 1]
    for i in range(2, n):
        seq.append(seq[-1] + seq[-2])
    return seq[:n]

fib_seq = fibonacci(5)
print("\nFibonacci Projection Sequence (5 years):", fib_seq)
```

```python
# Function to project next 5 years using Fibonacci increments
def project_population(data, fib_seq):
    last_val = data[-1]
    projections = [last_val + f*10 for f in fib_seq]  # scaling growth by 10 for realism
    return np.array(projections)

kampala_proj = project_population(kampala, fib_seq)
wakiso_proj  = project_population(wakiso, fib_seq)
gulu_proj    = project_population(gulu, fib_seq)

print("\nProjected Populations (next 5 years):")
print("Kampala:", kampala_proj)
print("Wakiso:", wakiso_proj)
print("Gulu:", gulu_proj)

# Compare variance of actual vs projected
print("\nVariance Comparison:")
print("Kampala - Actual:", np.var(kampala), "Projected:", np.var(kampala_proj))
print("Wakiso  - Actual:", np.var(wakiso),  "Projected:", np.var(wakiso_proj))
print("Gulu    - Actual:", np.var(gulu),    "Projected:", np.var(gulu_proj))
```

```python
# Plot actual vs projected populations
years_actual = np.arange(1, 11)   # Years for actual data
years_proj   = np.arange(11, 16)  # Years for projections

plt.figure(figsize=(10,6))

# Kampala
plt.plot(years_actual, kampala, 'bo-', label="Kampala Actual")
plt.plot(years_proj, kampala_proj, 'b--', label="Kampala Projected")

# Wakiso
plt.plot(years_actual, wakiso, 'go-', label="Wakiso Actual")
plt.plot(years_proj, wakiso_proj, 'g--', label="Wakiso Projected")

# Gulu
plt.plot(years_actual, gulu, 'ro-', label="Gulu Actual")
plt.plot(years_proj, gulu_proj, 'r--', label="Gulu Projected")

plt.title("UBOS Multi-District Population Forecast")
plt.xlabel("Year")
plt.ylabel("Population (thousands)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Solar Micro-Grid Simulation
# Question 2 Solution
# ---------------------------------------------
# Author: [Your Name]
# Date: [Submission Date]

import numpy as np
from scipy.linalg import solve
import statistics as stats
import matplotlib.pyplot as plt
```

```python
# Matrix coefficients from the equations:
# 3x + 2y = D1
# 4x + y  = D2
A = np.array([[3, 2],
              [4, 1]])

# Store results
results = []

print("Enter daily demand values for 7 days (D1 and D2):")
for day in range(1, 8):
    D1 = float(input(f"Day {day} - Enter D1: "))
    D2 = float(input(f"Day {day} - Enter D2: "))
    
    # Solve system of equations
    b = np.array([D1, D2])
    x, y = solve(A, b)   # x = solar panels, y = batteries
    
    results.append([x, y])

# Convert results to NumPy array
results = np.array(results)
print("\nDaily Solar & Battery Usage (x, y):\n", results)
```

```python
# 2.4 Separate solar (x) and battery (y) usage
solar_usage = results[:, 0]
battery_usage = results[:, 1]

print("\nSolar Usage Statistics:")
print("Mean:", stats.mean(solar_usage))
print("Variance:", stats.variance(solar_usage))
print("Standard Deviation:", stats.stdev(solar_usage))

print("\nBattery Usage Statistics:")
print("Mean:", stats.mean(battery_usage))
print("Variance:", stats.variance(battery_usage))
print("Standard Deviation:", stats.stdev(battery_usage))
```

```python
# 2.5 Plot daily solar vs battery usage
days = np.arange(1, 8)

plt.figure(figsize=(8,5))
plt.plot(days, solar_usage, 'bo-', label="Solar Panels (x)")
plt.plot(days, battery_usage, 'ro-', label="Batteries (y)")

plt.title("Daily Solar vs Battery Usage (7 Days)")
plt.xlabel("Day")
plt.ylabel("Energy Contribution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Lake Victoria Fish Export Risk Model
# Question 3 Solution



import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
```

```python
# 3.1 Fibonacci generator
def fibonacci(n):
    seq = [0, 1]
    for i in range(2, n):
        seq.append(seq[-1] + seq[-2])
    return seq[:n]

# Fish stock growth (15 days)
fish_stock = np.array(fibonacci(15))
print("Fish Stock (Fibonacci sequence):", fish_stock)
```

```python
# 3.2 Simulated fish prices (UGX per kg) for 15 days
fish_prices = np.array([12000, 12500, 11800, 13000, 12800, 
                        13200, 12700, 13500, 14000, 13800, 
                        13600, 14200, 14500, 15000, 14800])

print("Fish Prices (UGX per kg):", fish_prices)
```

```python
# 3.3 Daily revenue = stock * price
revenue = fish_stock * fish_prices
print("\nDaily Revenue (UGX):", revenue)
```

```python
# 3.4 
mean_rev = stats.mean(revenue)
median_rev = stats.median(revenue)
variance_rev = stats.variance(revenue)

print("\nRevenue Statistics:")
print("Mean:", mean_rev)
print("Median:", median_rev)
print("Variance:", variance_rev)
```

```python
# 3.5
 
if variance_rev > 50000:
    print("Risk Level: HIGH RISK ")
else:
    print("Risk Level: LOW RISK ")

```

```python
# 3.6 Plot fish stock vs revenue trend
days = np.arange(1, 16)

plt.figure(figsize=(10,5))

# Plot stock
plt.plot(days, fish_stock, 'bo-', label="Fish Stock (kg)")

# Plot revenue
plt.plot(days, revenue, 'r^-', label="Revenue (UGX)")

plt.title("Lake Victoria Fish Export Risk Model")
plt.xlabel("Day")
plt.ylabel("Stock & Revenue")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Weather & Agriculture Analysis
# Question 4 Solution
# ------------------------------------------------

import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
import math
```

```python
# Rainfall data (mm) for Kampala across 12 months
kampala_rainfall = np.array([120, 140, 180, 200, 220, 180, 
                             90, 70, 60, 100, 110, 130])
print("Kampala Rainfall (mm):", kampala_rainfall)
```

```python
# 4.2
mean_kampala = stats.mean(kampala_rainfall)
print("\nMean Rainfall in Kampala:", mean_kampala)

# Classification
classification = ["Good for maize" if val > 150 else "Drought risk" 
                  for val in kampala_rainfall]

print("\nMonthly Classification:")
for i, status in enumerate(classification, start=1):
    print(f"Month {i}: {status}")
```

```python
# 4.3
# Example simulated rainfall data for Gulu
gulu_rainfall = np.array([100, 120, 160, 210, 190, 170, 
                          95, 80, 75, 105, 115, 125])
print("\nGulu Rainfall (mm):", gulu_rainfall)

```

```python
# 4.4 
# Dot product
dot_product = np.dot(kampala_rainfall, gulu_rainfall)

# Magnitudes
mag_kampala = np.linalg.norm(kampala_rainfall)
mag_gulu = np.linalg.norm(gulu_rainfall)

# Cosine similarity
cos_sim = dot_product / (mag_kampala * mag_gulu)

print("\nCosine Similarity (Kampala vs Gulu rainfall):", cos_sim)
```

```python
# 4.5
months = np.arange(1, 13)

plt.figure(figsize=(10,5))

plt.plot(months, kampala_rainfall, 'bo-', label="Kampala Rainfall")
plt.plot(months, gulu_rainfall, 'go-', label="Gulu Rainfall")

plt.title("Rainfall Comparison: Kampala vs Gulu")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Taxi Transport Revenue & Variability
# Question 5 Solution



import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
from scipy.linalg import solve
```

```python
# 5.1 Passenger counts for 10 days
passengers = np.array([35, 40, 42, 50, 55, 60, 48, 52, 47, 45])
print("Passenger Counts:", passengers)
```

```python
# 5.2
fare_per_passenger = 2000
revenues = passengers * fare_per_passenger
print("\nDaily Revenues (UGX):", revenues)
```

```python
# 5.3
mean_rev = stats.mean(revenues)
variance_rev = stats.variance(revenues)
stdev_rev = stats.stdev(revenues)

print("\nRevenue Statistics:")
print("Mean Revenue:", mean_rev)
print("Variance:", variance_rev)
print("Standard Deviation:", stdev_rev)
```

```python
# 5.4
# Coefficient matrix
A = np.array([[2, 1],
              [1, 3]])

# Example demands (can be adjusted)
D1, D2 = 120, 200
b = np.array([D1, D2])

solution = solve(A, b)
taxis, trips_per_taxi = solution

print("\nSupply-Demand System Solution:")
print("Taxis needed:", taxis)
print("Trips per taxi:", trips_per_taxi)
```

```python
# 5.5
forecast_rev = np.mean(revenues[-3:])
print("\nForecasted Revenue for Day 11:", forecast_rev)
```

```python
# 5.6
days_actual = np.arange(1, 11)
days_forecast = np.array([11])

plt.figure(figsize=(10,5))

# Plot actual revenues
plt.plot(days_actual, revenues, 'bo-', label="Actual Revenues")

# Plot forecasted revenue
plt.plot(days_forecast, [forecast_rev], 'r^', markersize=10, label="Forecasted Revenue")

plt.title("Taxi Transport Revenues: Actual vs Forecasted")
plt.xlabel("Day")
plt.ylabel("Revenue (UGX)")
plt.legend()
plt.grid(True)
plt.show()
```
