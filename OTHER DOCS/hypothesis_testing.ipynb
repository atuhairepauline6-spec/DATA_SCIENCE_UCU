In-class Exercise: Using the diamonds dataset, test the hypothesis that the average price of all carat categories
are the same.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
diamonds=pd.read_excel("diamonds_new.xlsx")

print(diamonds.head())
```

```python
x = diamonds['x']
y = diamonds['y']

t_stat, p_value = stats.ttest_ind(x,y,equal_var = False)
#print results
print("T-statistic:", t_stat)
print("P-value:", p_value)
print(np.mean(x), np.mean(y))


if p_value < 0.05:
  print(" We can reject the null hypothesis")
else:
  print("We can accept the null hypothesis")
```

i. Null Hypothesis: μ = M
ii. Alternative Hypothesis: μ ≠ M

Interpretation of the results:

The p-value is greater than 0.05(signifance level) and the mean of x is not equal to the mean of y
We reject the null hypothesis
