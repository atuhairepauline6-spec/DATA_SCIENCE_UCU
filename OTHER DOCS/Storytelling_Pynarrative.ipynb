```python
# Create a sample dataset: Yearly fish sales in million dollars
data = pd.DataFrame({
    'Year': [2000, 2005, 2010, 2015, 2020],  # Years of observation
    'Tilapia Sales (Million $)': [200, 450, 700, 500, 850]  # Sales data in million dollars
})
#View dataframe
data
```

```python
# Create a base visualization using PyNarrative
tilapia = (pn.Story(data)
                 .mark_line(color='firebrick')
                 .encode(x='Year:O', y='Tilapia Sales (Million $):Q')
                 .add_title("Tilapia Sales From 2000-2020",
                            "Yearly trend",
                            title_color="#b22222")
                 .add_context("Sales declined in 2015", position='top right')
                 .add_annotation('2020', 1000, "Peak in 2020", arrow_color='gray', label_color='black')
                 .render())
tilapia
```

```python
#Create a barchart
tilapia_bar = (pn.Story(data, width=600)
                 .mark_bar(color='teal')
                 .encode(x='Year:O', y='Tilapia Sales (Million $):Q')
                 .add_title("Tilapia Sales in Uganda", "2000-2020",
                            title_color="#333")
                 .add_context("Gradual Growth in Sales", position='top')
                 .add_annotation(2015, 500, "Fishing Regulation Impact", arrow_color='red', label_color='darkred')
                 .render())
tilapia_bar
```
